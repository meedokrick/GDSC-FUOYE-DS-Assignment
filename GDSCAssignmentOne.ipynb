```python
import pandas as pd

#Data collection
#reading the data collected
data = pd.read_csv('NY2017.csv')
data

#Data cleaning
#checking the proportion of missing data on each column
print("Missing Data distribution")
print(data.isnull().mean())
print(data.isnull().sum())
print("")

#getting the datatype of each column
print("Columns Datatypes")
print(data.dtypes)

#dealing with missing values
#replacing object dtypes with None, and float dtypes with 0
missingObjColumns = ['Hospital Service Area','Hospital County','Zip Code - 3 digits','APR Severity of Illness Description','APR Risk of Mortality','Payment Typology 2','Payment Typology 3','Birth Weight']
missingFloatColumns = ['Operating Certificate Number','Permanent Facility Id']

for colOb in missingObjColumns:
    data[colOb] = data[colOb].fillna("None")
    
for colFl in missingFloatColumns:
    data[colFl] = data[colFl].fillna(0)
    
#Feature Engineering
#Getting lists of required features
#Dropping unneccessory columns to remain with the required column
inputData = data.drop(columns=['Hospital Service Area','Patient Disposition','Hospital County','Operating Certificate Number','Permanent Facility Id','Facility Name','Zip Code - 3 digits','Length of Stay','Discharge Year','Payment Typology 1','Payment Typology 2','Payment Typology 3','Birth Weight','Abortion Edit Indicator','Emergency Department Indicator','Total Charges','Total Costs'])
inputData

#Putting the output column into a separate column
outputData = data['Length of Stay']
outputData
```
